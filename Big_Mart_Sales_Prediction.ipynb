Handling Missing Values

Mean --> average

Mode --> more repeated value

```python
# mean value of "Item_Weight" column
big_mart_data['Item_Weight'].mean()
```

```python
big_mart_data['Item_Weight'] = big_mart_data['Item_Weight'].fillna(big_mart_data['Item_Weight'].mean())
```

```python
# mode of "Outlet_Size" column
big_mart_data['Outlet_Size'].mode()
```

```python
# filling the missing values in "Outlet_Size" column with Mode
mode_of_Outlet_size = big_mart_data.pivot_table(values='Outlet_Size', columns='Outlet_Type', aggfunc=(lambda x: x.mode()[0]))
```

```python
print(mode_of_Outlet_size)
```

```python
miss_values = big_mart_data['Outlet_Size'].isnull()
```

```python
print(miss_values)
```

```python
big_mart_data.loc[miss_values, 'Outlet_Size'] = big_mart_data.loc[miss_values,'Outlet_Type'].apply(lambda x: mode_of_Outlet_size[x])
```

```python
# checking for missing values
big_mart_data.isnull().sum()
```

Data Pre-Processing

```python
big_mart_data.head()
```

```python
big_mart_data['Item_Fat_Content'].value_counts()
```

```python
big_mart_data.replace({'Item_Fat_Content': {'low fat':'Low Fat','LF':'Low Fat', 'reg':'Regular'}}, inplace=True)
```

```python
big_mart_data['Item_Fat_Content'].value_counts()
```

Label Encoding

```python
encoder = LabelEncoder()
```

```python
big_mart_data['Item_Identifier'] = encoder.fit_transform(big_mart_data['Item_Identifier'])

big_mart_data['Item_Fat_Content'] = encoder.fit_transform(big_mart_data['Item_Fat_Content'])

big_mart_data['Item_Type'] = encoder.fit_transform(big_mart_data['Item_Type'])

big_mart_data['Outlet_Identifier'] = encoder.fit_transform(big_mart_data['Outlet_Identifier'])

big_mart_data['Outlet_Size'] = encoder.fit_transform(big_mart_data['Outlet_Size'])

big_mart_data['Outlet_Location_Type'] = encoder.fit_transform(big_mart_data['Outlet_Location_Type'])

big_mart_data['Outlet_Type'] = encoder.fit_transform(big_mart_data['Outlet_Type'])
```

```python
big_mart_data.head()
```

Splitting features and Target

```python
X = big_mart_data.drop(columns='Item_Outlet_Sales', axis=1)
Y = big_mart_data['Item_Outlet_Sales']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Machine Learning Model Training

XGBoost Regressor

```python
regressor = XGBRegressor()
```

```python
regressor.fit(X_train, Y_train)
```

Evaluation

```python
# prediction on training data
training_data_prediction = regressor.predict(X_train)

# R squared Value
r2_train = metrics.r2_score(Y_train, training_data_prediction)
print('R Squared value = ', r2_train)
```

```python
# prediction on test data
test_data_prediction = regressor.predict(X_test)

# R squared Value
r2_test = metrics.r2_score(Y_test, test_data_prediction)
print('R Squared value = ', r2_test)
```

```python
r2_train_xgb = metrics.r2_score(Y_train, regressor.predict(X_train))
r2_test_xgb  = metrics.r2_score(Y_test,  regressor.predict(X_test))
```

Linear Regression

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
# prediction on training data
training_data_prediction = lr.predict(X_train)

# R squared Value
r2_train = metrics.r2_score(Y_train, training_data_prediction)
print('R Squared value = ', r2_train)
```

```python
# prediction on test data
test_data_prediction = lr.predict(X_test)

# R squared Value
r2_test = metrics.r2_score(Y_test, test_data_prediction)
print('R Squared value = ', r2_test)
```

```python
r2_train_lr = metrics.r2_score(Y_train, lr.predict(X_train))
r2_test_lr  = metrics.r2_score(Y_test,  lr.predict(X_test))
```

Random Forest

```python
rf = RandomForestRegressor(n_estimators=100, random_state=2)
rf.fit(X_train, Y_train)
```

```python
# prediction on training data
training_data_prediction = rf.predict(X_train)

# R squared Value
r2_train = metrics.r2_score(Y_train, training_data_prediction)
print('R Squared value = ', r2_train)
```

```python
# prediction on test data
test_data_prediction = rf.predict(X_test)

# R squared Value
r2_test = metrics.r2_score(Y_test, test_data_prediction)
print('R Squared value = ', r2_test)
```

```python
r2_train_rf = metrics.r2_score(Y_train, rf.predict(X_train))
r2_test_rf  = metrics.r2_score(Y_test,  rf.predict(X_test))
```

```python
results = {
    'Model': ['Linear Regression', 'Random Forest', 'XGBoost'],
    'Train R²': [r2_train_lr, r2_train_rf, r2_train_xgb],
    'Test R²': [r2_test_lr, r2_test_rf, r2_test_xgb]
}
pd.DataFrame(results)
```
